```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv("dg_performance_2025.csv")

top10 = df.sort_values('total_true', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top10, x='total_true', y='player_name', palette='viridis')
plt.title('Top 10 Players by Total Strokes Gained')
plt.xlabel('Total Strokes Gained')
plt.ylabel('Player')
plt.tight_layout()
plt.show()
```
